```python
import os
import torch

from ultralytics import YOLO
from torch import cuda
from PIL import Image
```

```python
device = f"`cuda:{cuda.current_device()}" if cuda.is_available() else "cpu"
print(device)
```

```python
model = YOLO("yolov5n.pt")
```

```python
model.to('cuda')
model.train(data="train.yaml", epochs=10, imgsz=256, batch=32)
```

```python
from ultralytics.utils.metrics import DetMetrics

metrics = DetMetrics()
print(dir(metrics))  # Menampilkan semua atribut dan metode yang tersedia
```

```python
from ultralytics.utils.metrics import DetMetrics

metrics = DetMetrics()
print(metrics.results_dict)  # Periksa hasil evaluasi sebelum curves_results
print(metrics.curves_results)  # Coba akses setelah memastikan evaluasi dilakukan
```
