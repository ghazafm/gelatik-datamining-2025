```python
import os
from roboflow import Roboflow

# Initialize the Roboflow object with your API key
rf = Roboflow(api_key="IK9Jb9qlny3R7ROyGSnr")

# Define the workspace and project details
workspace_id = "fadhilah-hilmi-hym4y"
project_id = "bcc-telat-daftar"
project = rf.workspace(workspace_id).project(project_id)

# Define the folder containing your images
folder_path = "/content/254p RGB Images"

# Iterate through each file in the folder and upload it
for filename in os.listdir(folder_path):
    file_path = os.path.join(folder_path, filename)
    if os.path.isfile(file_path) and filename.lower().endswith(('.png', '.jpg', '.jpeg')):
        print(f"Uploading {filename}...")
        project.upload(image_path=file_path)
        print(f"Uploaded {filename}")
```
